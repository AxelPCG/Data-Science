```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14, 6))
```

```python
dados = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
dados
```

```python
valor = dados['Valor m2']
```

```python
Q1 = valor.quantile(.25).round(2)
Q3 = valor.quantile(.75).round(2)
IIQ = (Q3 - Q1).round(2)
limite_inferior = (Q1 - 1.5 * IIQ).round(2)
limite_superior = (Q3 + 1.5 * IIQ).round(2)
```

```python
print('[Q1] -> {}'.format(Q1))
print('[Q3] -> {}'.format(Q3))
print('[IIQ] -> {}'.format(IIQ))
print('[limite inferior] -> {}'.format(limite_inferior))
print('[limite superior] -> {}'.format(limite_superior))
```
